```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

data collection and data processing


```python
sonar_data=pd.read_csv('/content/sonar data.csv')
```

```python
sonar_data.head()
```

```python
#no of rows and coloumns
sonar_data.shape
```

```python
sonar_data.describe()
```

```python
# Access the 60th column using iloc. Assuming index starts at 0, this is the 61st column.
sonar_data.iloc[:,60].value_counts()
```

```python
# Group by the 60th column (index 60).
sonar_data.groupby(sonar_data.columns[60]).mean()
```

```python
#seperate the data and labels
x=sonar_data.drop(columns=sonar_data.columns[60],axis=1)
y=sonar_data[sonar_data.columns[60]]
```

```python
print(x)
```

```python
print(y)
```

```python
#training and test data
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.1,stratify=y,random_state=1)
```

```python
print(x.shape,x_train.shape,x_test.shape,y_train.shape,y_test.shape)
```

```python
# model training -logistical model
model=LogisticRegression()
```

```python
model.fit(x_train,y_train)
```

```python
# accuracy an training data
x_train_prediction=model.predict(x_train)
training_data_accuracy=accuracy_score(x_train_prediction,y_train)
```

```python
print('Accuracy on training data:',training_data_accuracy)
```

```python
x_test_prediction=model.predict(x_test)
testing_data_accuracy=accuracy_score(x_test_prediction,y_test)
```

```python
print('Accuracy on testing data:',testing_data_accuracy)
```
